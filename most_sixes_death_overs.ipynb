```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('deliveries.csv')
```

```python
data.head()
```

```python
# problem 1 : top 5 batsman with most no. of sixes in last 5 overs 

bat=data.groupby('batsman_runs')

# we take out the group of data where batsman run is 6

six=bat.get_group(6)

# this is how the group_data looks like
six
```

```python
# here we mask the overs to be greater than or equal  to 15  
# this returns 6 runs data for last 5 overs and then we count values of batsman

six[six['over']>=15]['batsman'].value_counts().sort_values(ascending=False).head()
```

```python
#problem 2 : top 5 bowlers against whom V kohli has scored most runs
# masking dataframe with batsman as V kohli

vk=data[data['batsman']=='V Kohli']

#this is how dataset looks after masking

vk
```

```python
# for top 5 bowlers against whom kohli scored most runs

vk.groupby('bowler')['batsman_runs'].sum().sort_values(ascending=False).head()
```
