```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
# merging two datasets
full_match=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
# problem 1

def runs_plot(batsman_name):
    # groupby data according to season and batsman then creating a new dataframe
    
    required_data=full_match.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
    
    # selecting a batsman from dataframe
    
    required_data[required_data['batsman']==batsman_name]
    
    
    x=required_data[required_data['batsman']==batsman_name]['season'].tolist()
    y=required_data[required_data['batsman']==batsman_name]['batsman_runs'].tolist()
    
    # plotting the graph
    
    plt.xlabel('season')
    plt.ylabel('batsman runs')
    plt.title('batsman runs and season comparison')
    plt.plot(x,y,marker='D') 
    plt.show()
    
```

```python
# testing the function
runs_plot("V Kohli")
```

```python
# problem 2
def pie_runs(batsman_name):
    
    # finding a dataset for required batsman
    runs=full_match[(full_match['batsman']==batsman_name) & (full_match['batsman_runs']!=0)]
    
    # finding total_runs scored in different modes
    x=runs.groupby(['batsman_runs'])['batsman_runs'].sum().tolist()
    y=runs.groupby(['batsman_runs'])['batsman_runs'].sum().index.tolist()
    
    # plotting the graph
    plt.pie(x,labels=y,autopct='%0.1f%%',shadow=True)
    plt.show()

    
```

```python
# testing the code
pie_runs('V Kohli')
```

```python
pie_runs('MS Dhoni')
```
